```python
#Problem #2
marks = [45, 78, 90, 33, 60]
passing_mark = 50

# Counting passing and failing students
passed = [m for m in marks if m >= passing_mark]
failed = [m for m in marks if m < passing_mark]

# Displaying results
print(f"Total Students: {len(marks)}")
print(f"Passed Students ({len(passed)}): {passed}")
print(f"Failed Students ({len(failed)}): {failed}")
print("-" * 20)
print(f"Final Count: {len(passed)} Pass, {len(failed)} Fail")
```

```python
#Problem #4
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")

    if length > 10:
        print("  → This message is longer than 10 characters.")
```
